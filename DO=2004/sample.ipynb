```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_file = 'data/feedback-types-per-order.csv'  # Replace with your actual file name
df = pd.read_csv(csv_file)

#df = df.sort_values(by='aufnr')

# Melt the DataFrame for seaborn (long format)
#df = df[["aufnr", "rm_l", "rm_m", "rm_n"]]
melted_df = df.melt(id_vars='aufnr', var_name='count_type', value_name='count')

# Plot with seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=melted_df, x='aufnr', y='count')

# Styling
plt.title('Order Counts Distribution')
plt.xlabel('Order ID')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.legend(title='Count Type')

# Show it
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data simulating your CSV
df = pd.read_csv(csv_file)
df = df.apply(pd.to_numeric)
df = df[["aufnr", "rm_l", "rm_m", "rm_n"]]
# Melt the DataFrame for seaborn
melted_df = df.melt(id_vars='aufnr', var_name='count_type', value_name='count')

# Plot as line chart
plt.figure(figsize=(20, 6))
sns.lineplot(data=melted_df, x='aufnr', y='count', hue='count_type')

# Style it
plt.title('Order Counts Distribution')
plt.xlabel('Order ID')
plt.ylabel('Count')
plt.grid(True)
plt.legend(title='Count Type')
plt.tight_layout()
plt.show()
```
